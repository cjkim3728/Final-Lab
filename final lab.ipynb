# Question 1: Check the membership of f_a

$$f_a(x) =  \begin{array}{ll}
x^a & \textrm{for } x>0 \\
0 & \textrm{for } x \leq 0 \\
\end{array}
$$ where $a>0$

First, check the continuity of $f_a$.

Since $\lim_{x\rightarrow 0^-}f_a(x)=\lim_{x\rightarrow 0^+}f_a(x)=0$, $\lim_{x\rightarrow 0}f_a(x)$ exists and equals to 0. So, $0=f_a(0)=\lim_{x\rightarrow 0}f_a(x)$ and $f_a(x)$ is continuous on $[-1,1]$

For $x>0$ and $k \in \mathbb{N}$, $k$-th derivative of $f_a$ is $f_a^{(k)}(x)=a(a-1)\ldots(a-k)x^{a-k}$. 

If $a>k$, $$\lim_{x\rightarrow 0^+}f_a^{(k)}(x)=\lim_{x\rightarrow 0^+}a(a-1)\ldots(a-k)x^{a-k}=0=\lim_{x\rightarrow 0^-}f_a^{(k)}(x)$$, and just like $f_a(x)$ case, $f_a^{(k)}(x)$ is continuous on $[-1,1]$.

However, if $a\leq k$, $$\lim_{x\rightarrow 0^+}x^{a-k}=\lim_{x\rightarrow 0^+}\frac{1}{x^{k-a}}=\frac{1}{0^{k-a}}$$, and $\lim_{x\rightarrow 0^-}f_a^{(k)}(x) \neq \lim_{x\rightarrow 0^+}f_a^{(k)}(x)$. So, there is a discontinuity at $x=0$ for $a\leq k$, which implies that $f_a^{(k)}(x)$ is only continuous for $k \in \mathbb{N}$ and $k<a$.

Thus, findng maximum $k$ is equivalent to finding largest integer $k$ that is less than $a$.
$$k = \begin{array}{ll}
a-1 & \textrm{for } a \in \mathbb{N} \\
\lfloor a \rfloor & \textrm{else}\\
\end{array}$$

Consider $a \in \mathbb{N}$. Then, $k=a-1$ and $$f_a^{(k)}(x)= \begin{array}{ll}
a!x & \textrm{for } x>0 \\
0 & \textrm{for } x \leq 0 \\
\end{array}$$ 

and for $-1 \leq x < y \leq 1$

$$\sup_{-1\leq x<y \leq 1} \frac {|f(x)-f(y)|}{|x-y|^\alpha} = \sup_{-1\leq x<y \leq 1} \frac {a!|x-y|}{|x-y|^\alpha}\leq \sup_{-1\leq x<y \leq 1} a!|x-y|^{1-\alpha}$$

So, for $\alpha$ that is $0<\alpha \leq 1$, $$\sup_{-1\leq x<y \leq 1} a!|x-y|^{1-\alpha} \leq 2a!<\infty$$ and $f^{(k)} \in C^{0,\alpha}([-1,1])$


Consider $a \in \mathbb{R} \setminus \mathbb{N}$. Then, $k=\lfloor a \rfloor$ and $$f_a^{(k)}(x)= \begin{array}{ll}
\lfloor a \rfloor!x^{frac(a)} & \textrm{for } x>0 \\
0 & \textrm{for } x \leq 0 \\
\end{array}$$ Lets's denote $frac(a)$ as $\beta$ for convinience. Then, for $-1 \leq x < y \leq 1$ 
$$\sup_{-1\leq x,y \leq 1} \frac {|f(x)-f(y)|}{|x-y|^\alpha} = \sup_{-1\leq x,y \leq 1} \frac {\lfloor a \rfloor!|x^\beta-y^\beta|}{|x-y|^\alpha}\leq \sup_{-1\leq x,y \leq 1} \frac {\lfloor a \rfloor!|x-y|^\beta}{|x-y|^\alpha} = \sup_{-1\leq x,y \leq 1} \lfloor a \rfloor!|x-y|^{\beta-\alpha}$$

Then, for $0<\alpha \leq \beta < 1$ $$\sup_{-1\leq x,y \leq 1} \lfloor a \rfloor!|x-y|^{\beta-\alpha} \leq \lfloor a \rfloor!2^{\beta-\alpha}<\infty$$ and $f^{(k)} \in C^{0,\alpha}([-1,1])$

However, for $0<\beta < \alpha < 1$ there exists $r=\alpha-\beta$ such that $0<r<1$ and $$\sup_{-1\leq x,y \leq 1} \lfloor a \rfloor!|x-y|^{\beta-\alpha} = \sup_{-1\leq x,y \leq 1} \lfloor a \rfloor!|x-y|^{-r} = \sup_{-1\leq x,y \leq 1} \frac {\lfloor a \rfloor!}{|x-y|^r}$$
For $x \rightarrow y$, $\lim_{x\rightarrow y}\frac {\lfloor a \rfloor!}{|x-y|^r}=\frac {\lfloor a \rfloor!}{0}$, which implies limit does not exist. Since there is no limit of $\frac {\lfloor a \rfloor!}{|x-y|^r}$, $\sup_{-1\leq x,y \leq 1} \frac {\lfloor a \rfloor!}{|x-y|^r}$ does not exist as well.

Thus, for $0<\beta < \alpha < 1$, $f^{(k)} \notin C^{0,\alpha}([-1,1])$.

In summary, $f \in C^{k}([-1,1])$ and $f^{(k)} \in C^{0,\alpha}([-1,1])$ for $k \in \mathbb{N}$, $k<a$ and $0<\alpha \leq \lfloor a \rfloor$.

# Question 2: Piecewise Constant Function

First, code $f_a(x)$, where 
$$f_a(x) =  \begin{array}{ll}
x^a & \textrm{for } x>0 \\
0 & \textrm{for } x \leq 0 \\
\end{array}
$$



```python
def f(x,a):
    if (x > 0):
        return x**a
    return 0
```

Now, we will try to approximate $f_a$ by a piecewise constant function on the uniform partition of $[-1,1]$ consisting of n subintervals. For each subinterval, we will use the value of $f_a$ at the midpoint as the approximation.

```python
def constantF(x, a, n):
    # divide (-1,1) into n subintervals
    # if x is in one of the interval, find the midpoint of that interval
    # return f(midpoint,a)
    if x<=-1:
        return 0
    else:
        h = 2/n
        for i in range(1,n+1):
            if (-1+i*h > x):
                mid = -1+(2*i-1)/2*h
                return f(mid, a)
        return f(1,a)
```

### 2) Computing error of the approximation in the maximum norm

Maximum norm of error: 
$$||f_a(x)-E_a(x)||_\infty := \max_{x \in [-1,1]} |f_a(x)-E_a(x)|$$
where $E_a(x)$ is the approximated function of $f_a$ we derived from above.

```python
def diff(x,a,n):
    # return the absolute difference of f_a(x) and E_a(x) at x
    return np.abs(f(x,a) - constantF(x,a,n))
```

```python
def normDiff(xlist,a,n):
    max = 0
    for x in xlist:
        if (max < diff(x,a,n)):
            max = diff(x,a,n)
    return max
```

### 3) Estimate the rate of convergence

For $n$ large, we expect the error of the approximation to behave like $n^{-r}$ for some $r$, and we call $r$,  the rate of convergence. We will try to estimate $r$ for a different values of $a$, focusing around the point $a=1$. 

```python
xarray = np.linspace(-1,1,2000)
nlist = np.array(range(5,200,6))

fig = plt.figure(figsize=(8,24))

alist= 1.1, 0.9, 1.2, 0.8, 1.3, 0.7, 1.4, 0.6, 1.5, 0.5, 1
k = 1
for a_i in alist:
    ylist = []
    for n in nlist:
        ylist.append(normDiff(xarray,a_i,n))
    plt.subplot(6,2,k)
    plt.plot(nlist,ylist, color = 'orange')
    plt.title('a = %s' % a_i)
    k=k+1

plt.show()
```

If the error of the approximation behaves as $n^{-r}$, then $y = cn^{-r}$ for some constant c. We can estimate r by taking log at each side of the equality.
$$\log(y)=\log(cn^{-r})=-r\log(n)+C$$
for some constant $C$.
If we plot $log(y)$ vs $log(n)$, the slope of this linear plot will be $-r$.

# Question 3: Continuous Piecewise Linear Function

This time, we will try to approximate $f_a$ by a piecewise linear function on the uniform partition of $[-1,1]$ consisting of n subintervals.

```python
def lin(x,start,end,a):
    #linear function crossing (start,f(start,a)) and (end,f(end,a))
    # return the value of that linear function at x where x is between start and end
    slope = (f(end,a)-f(start,a))/(end-start)
    return slope*(x-start)+f(start,a)
```

```python
def linearF(x, a, n):
    # divide (-1,1) into n subintervals
    # locate which interval x is in
    # draw a line across two endpoints of that interval
    # return value of that linear function at x
    if x<=-1:
        return 0
    else:
        h = 2/n
        for i in range(1,n+1):
            if (-1+i*h > x):
                return lin(x,-1+(i-1)*h,-1+i*h,a)
        return f(1,a)
```

# Question 4: Bernstein Polynomial of degree n 

Bernstein polynomial:
$$\beta_{n,k}(x)=\binom{n}{k}(1-x)^{n-k}x^k$$
and approximate of $f_a$ with it is
$$B_a(x)=\sum_{k=0}^n f_a \left(\frac{k}{n}\right)\beta_{n,k}(x)$$

```python
def bernsteinPoly (n,k,x):
    return sc.binom(n,k)*(1-x)**(n-k)*x**k
```

```python
def bernsteinF(x, a, n):
    sum = 0
    h = 2/n
    for k in range(0, n+1):
        sum = sum + f(-1+k*h,a)*bernsteinPoly(n,k,(x+1)/2)
    return sum
```

# Question 5: Chebyshev interpolation Polynomial of degree n 

The Chebyshev nodes in the interval $[-1,1]$ are defined as
$$x_k = \cos \left(\frac{2k-1}{2n}\pi\right), \ k=1,\ldots,n.$$
And approximation of $f_a(x)$ using Lagrange interpolation polynomial is 
$$L(x) := \sum_{k=0}^nf_a(x_k)l_k(x)$$
where
$$l_k(x) := \prod_{\begin{array}{l}i=0 \\i \neq k\\\end{array} }^n \frac{x-x_i}{x_k-x_i}$$

```python
def chebyNodes(n):
    
    nodes = []
    for i in range (1,n+1):
        nodes.append(np.cos((2*i-1)/(2*n)*np.pi))
    return nodes
```

```python
def lagrange(nodes,k,x):
    mult=1
    for i in range (0,len(nodes)):
        if i != k:
            mult = mult * (x-nodes[i])/(nodes[k]-nodes[i])
    return mult
```

```python
def chebyLag(x,a,n):
    nodes = chebyNodes(n)
    sum = 0
    for k in range (0,n):
        sum = sum + lagrange(nodes,k,x)*f(nodes[k],a)
    return sum
```

```python
def normDiff(xlist,a,n):
    max = 0
    for x in xlist:
        if (max < diff(x,a,n)):
            max = diff(x,a,n)
    return max
```

```python
xarray = np.linspace(-1,1,1000)
normDiff(xarray,2,5)
```
